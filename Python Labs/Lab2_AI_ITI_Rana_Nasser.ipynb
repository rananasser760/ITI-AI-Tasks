```python
''''2- Convert two lists into a dictionary
 	keys = ['Ten', 'Twenty', 'Thirty']
 	values = [10, 20, 30]
'''
keys = ['Ten', 'Twenty', 'Thirty']
values = [10, 20, 30]
dec=dict(zip(keys,values))
print("dictionary: ",dec)
```
